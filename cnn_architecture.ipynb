```python
#5
import numpy as np
import tensorflow as tf

# Create random data
x_train = np.random.rand(100, 28, 28, 1)
y_train = np.random.randint(0, 10, 100)

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
```

```python
#THEORY
# 1
# A Convolutional Neural Network (CNN) is a type of deep learning model designed to process structured grid data, such as images. It is particularly effective for image classification, object detection, and other tasks related to visual data. CNNs use convolutional layers that automatically detect patterns like edges, textures, and shapes, making them ideal for image processing.
```

```python
# The key components of a CNN architecture include:

# Convolutional layers: Apply filters to input data to create feature maps.
# Activation functions (usually ReLU): Introduce non-linearity to the model.
# Pooling layers (max pooling or average pooling): Downsample feature maps to reduce their dimensionality.
# Fully connected layers: Perform high-level reasoning using the features learned in earlier layers.
# Softmax layer: Typically used in classification tasks to output probabilities.
```

```python

# The convolutional layer applies filters (kernels) to the input image to extract features. It slides the filter across the image, performing element-wise multiplication and summing the results to produce a feature map. These feature maps help detect patterns such as edges, textures, or specific objects in the image.
```

```python
# A filter (also called a kernel) is a small matrix used to convolve an image. It contains learnable weights that allow the network to detect specific features such as edges, corners, or textures. Filters are learned during the training process and help capture the spatial hierarchy of features in an image.
```

```python
# Pooling is a downsampling operation that reduces the spatial dimensions (height and width) of feature maps. It helps reduce computation, prevent overfitting, and make the model invariant to small translations in the input. Pooling layers typically use operations like max pooling or average pooling.
```

```python
# Max pooling: Selects the maximum value from each patch of the feature map (most common).
# Average pooling: Takes the average value from each patch.
# Global pooling: Computes a single value (such as max or average) for the entire feature map, typically used before the fully connected layers.
```

```python
# Backpropagation in CNNs works similarly to other neural networks. It involves the following steps:

# Forward pass: The input is passed through the layers, and the output is computed.
# Loss calculation: The error (loss) between the predicted output and the true labels is calculated.
# Backward pass: The gradients of the loss with respect to the weights are calculated by applying the chain rule. These gradients are propagated back through the network.
# Weight update: The weights are adjusted using optimization algorithms like gradient descent.
```

```python
# Activation functions introduce non-linearity into the model, allowing it to learn complex patterns. Without activation functions, the network would only be able to learn linear relationships. The most common activation function in CNNs is ReLU (Rectified Linear Unit), which helps with faster convergence and reduces the vanishing gradient problem.
```

```python
# A receptive field refers to the region of the input image that a particular neuron or filter in the network is "looking at." As you go deeper into the network, the receptive field increases, allowing the model to learn higher-level features that are spatially larger
```

```python
# Tensor space refers to the multidimensional arrays (tensors) that are used to represent the data at each layer of a CNN. For example, an image might be represented as a 3D tensor (height, width, channels). Operations like convolutions and pooling manipulate these tensors, and the network learns to extract meaningful features from them.
```

```python
# LeNet-5 is one of the earliest CNN architectures, developed by Yann LeCun. It was designed for handwritten digit recognition (MNIST dataset). LeNet-5 demonstrated that CNNs could effectively learn hierarchical features from images, leading to the development of more complex architectures later on.
```

```python
# AlexNet, developed by Alex Krizhevsky, was a deep CNN architecture that won the 2012 ImageNet competition. It introduced the use of ReLU activation functions, data augmentation, and GPUs for training large models, which significantly improved performance on image classification tasks and demonstrated the power of deep learning.
```

```python
# VGGNet, developed by the Visual Geometry Group, is known for its simple and deep architecture, using only 3x3 convolution filters and 2x2 max pooling. Unlike AlexNet, which uses a mix of large convolutional filters, VGGNet's design is based on deep layers, making it more computationally expensive but very effective for large-scale image classification.
```

```python
# GoogLeNet, introduced by Google, is a deep CNN architecture that uses the Inception module, which includes filters of different sizes in parallel to capture features at multiple scales. The key innovation is the use of 1x1 convolutions to reduce the dimensionality of feature maps, making the network more efficient.
```

```python
# ResNet (Residual Networks) introduced the concept of skip connections or residual connections, which allow gradients to flow more easily during backpropagation. This solved the problem of vanishing gradients and enabled the training of very deep networks, such as those with hundreds of layers.
```

```python
# DenseNet (Densely Connected Convolutional Networks) is similar to ResNet but with denser connections between layers. In DenseNet, each layer receives input from all previous layers, allowing the network to reuse features more effectively. This dense connectivity leads to fewer parameters and better performance in some cases
```
